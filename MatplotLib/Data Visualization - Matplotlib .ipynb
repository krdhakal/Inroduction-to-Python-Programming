```python
import matplotlib.pyplot as plt
```

```python
year = [2015,2016,2017,2018,2019]
sales = [4356,4432,3333,2322,4323]
profit = [453,223,344,433,233]
```

```python
plt.plot(year,sales, 'red')
plt.plot(year, profit, 'blue')
plt.title('Year vs Profit and Sales analysis')
plt.xlabel('Years')
plt.ylabel('Profit and Sales')
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(year, sales, 'red')
plt.xlabel('Years')
plt.ylabel('Sales')
plt.title('Sales by Year')

plt.subplot(2,2,2)
plt.plot(year, profit, 'blue')
plt.xlabel('Years')
plt.ylabel('Profit')
plt.title('Profit by Year')
plt.tight_layout()
```

```python
import os
os.chdir(r'D:\Courses\Python Basics to Advanced')
import pandas as pd
```

```python
car = pd.read_csv('Car_sales.csv')
```

```python
for group, data in car.groupby('Manufacturer'):
    print(group,data['4-year resale value'].mean())
```

```python
plt.figure(figsize=(10,7))
for group, data in car.groupby('Manufacturer'):
    plt.barh(group,data['4-year resale value'].mean())
```

```python
car['Vehicle type'].unique()
```

```python
vt_dict = dict()
for i in set(car['Vehicle type']):
    vt_dict[i] = car[car['Vehicle type']==i].Manufacturer.count()
vt_dict
plt.bar(vt_dict.keys(), vt_dict.values())
```

```python
plt.figure(figsize=(10,8))
for group, data in car.groupby('Manufacturer'):
    plt.barh(group, data['4-year resale value'].mean())
    plt.barh(group, data['Price in thousands'].mean(), left = data['4-year resale value'].mean())
```
